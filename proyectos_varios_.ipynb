```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de Z
z = np.linspace(0, 1, 100)

# Calcular la función de densidad de probabilidad (PDF) de Z
pdf_z = np.where((z >= 0) & (z <= 1), 2 * (1 - z), 0)

# Calcular la función de distribución acumulativa (CDF) de Z
cdf_z = np.where(z < 0, 0, np.where(z <= 1, (1 - z) ** 2, 1))

# Graficar la PDF de Z
plt.subplot(1, 2, 1)
plt.plot(z, pdf_z, 'b-', linewidth=2)
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.title('PDF de Z = U - V')
plt.grid(True)

# Graficar la CDF de Z
plt.subplot(1, 2, 2)
plt.plot(z, cdf_z, 'r-', linewidth=2)
plt.xlabel('Valor de Z')
plt.ylabel('Probabilidad')
plt.title('CDF de Z = U - V')
plt.grid(True)

# Ajustar el diseño de los subplots
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = 0  # Parámetro a de la función
b = 0.05  # Parámetro b de la función
theta = np.linspace(0, 12 * np.pi, 1000)  # Valores de theta

r = a + b * theta  # Calcular los valores de r

x = r * np.cos(theta)  # Coordenadas x en función de r y theta
y = r * np.sin(theta)  # Coordenadas y en función de r y theta

# Calcular los límites del cuadrado
x_min, x_max = min(x), max(x)
y_min, y_max = min(y), max(y)
max_range = max(x_max - x_min, y_max - y_min)

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(4, 4), dpi=100)  # 10 cm x 10 cm

# Dibujar la espiral
ax.plot(x, y, color='blue')

# Añadir líneas radiales
num_radiales = 12  # Número de líneas radiales
angles = np.linspace(0, 2 * np.pi, num_radiales, endpoint=False)

for angle in angles:
    x_radial = np.linspace(-5, 5, 1000) * np.cos(angle)
    y_radial = np.linspace(-5, 5, 1000) * np.sin(angle)
    ax.plot(x_radial, y_radial, color='blue', alpha=0.5)

# Generar insectos aleatoriamente
num_insectos = 1000  # Número de insectos
insectos_x = np.random.uniform(x_min, x_min + max_range, num_insectos)
insectos_y = np.random.uniform(y_min, y_min + max_range, num_insectos)

# Verificar si los insectos chocan con la telaraña
insectos_chocados = 0

for i in range(num_insectos):
    distancia_espiral = np.sqrt((insectos_x[i] - x) ** 2 + (insectos_y[i] - y) ** 2)
    distancia_radios = np.min(np.sqrt((insectos_x[i] - x_radial) ** 2 + (insectos_y[i] - y_radial) ** 2))

    if np.any(distancia_espiral < 0.1) or distancia_radios < 0.1:
        ax.plot(insectos_x[i], insectos_y[i], 'ro')
        insectos_chocados += 1
    else:
        ax.plot(insectos_x[i], insectos_y[i], 'go')

# Configurar los límites del gráfico y el aspecto
ax.set_aspect('equal')
ax.set_xlim(x_min, x_min + max_range)
ax.set_ylim(y_min, y_min + max_range)
ax.axis('off')

# Mostrar el gráfico y el resultado de los insectos
plt.title("Insectos atrapados: {} - Insectos no atrapados: {}".format(insectos_chocados, num_insectos - insectos_chocados))
plt.show()
```

```python
import random
!pip install simpy
import simpy


RANDOM_SEED = 42
NEW_CUSTOMERS = 5  # Total number of customers
INTERVAL_CUSTOMERS = 10.0  # Generate new customers roughly every x seconds
MIN_PATIENCE = 1  # Min. customer patience
MAX_PATIENCE = 3  # Max. customer patience


def source(env, number, interval, counter):
    """Source generates customers randomly"""
    for i in range(number):
        c = customer(env, 'Customer%02d' % i, counter, time_in_bank=12.0)
        env.process(c)
        t = random.expovariate(1.0 / interval)
        yield env.timeout(t)


def customer(env, name, counter, time_in_bank):
    """Customer arrives, is served and leaves."""
    arrive = env.now
    print('%7.4f %s: Here I am' % (arrive, name))

    with counter.request() as req:
        patience = random.uniform(MIN_PATIENCE, MAX_PATIENCE)
        # Wait for the counter or abort at the end of our tether
        results = yield req | env.timeout(patience)

        wait = env.now - arrive

        if req in results:
            # We got to the counter
            print('%7.4f %s: Waited %6.3f' % (env.now, name, wait))

            tib = random.expovariate(1.0 / time_in_bank)
            yield env.timeout(tib)
            print('%7.4f %s: Finished' % (env.now, name))

        else:
            # We reneged
            print('%7.4f %s: RENEGED after %6.3f' % (env.now, name, wait))


# Setup and start the simulation
print('Bank renege')
random.seed(RANDOM_SEED)
env = simpy.Environment()

# Start processes and run
counter = simpy.Resource(env, capacity=1)
env.process(source(env, NEW_CUSTOMERS, INTERVAL_CUSTOMERS, counter))
env.run()
```

```python
!pip install simpy
import numpy as np
import matplotlib.pyplot as plt

def brownian_motion(T, N, seed=None):
    np.random.seed(seed)
    dt = T / float(N)
    dx = np.sqrt(dt) * np.random.randn(N)
    x = np.cumsum(dx)
    return x

T = 1.0     # Tiempo total
N = 1000    # Número de pasos
seed = 123  # Semilla para reproducibilidad

x = brownian_motion(T, N, seed)

t = np.linspace(0.0, T, N)

plt.plot(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.title('Movimiento Browniano')
plt.grid(True)
plt.show()
```

```python
!pip install simpy
import simpy
import random
import numpy as np
import matplotlib.pyplot as plt

class Autolavado:
    def __init__(self, env, num_servidores):
        self.env = env
        self.servidores = [simpy.Resource(env) for _ in range(num_servidores)]
        self.media_lavado = 5  # Media de tiempo de lavado en minutos
        self.tiempos_espera = []  # Lista para almacenar los tiempos de espera

    def lavar_auto(self, auto):
        tiempo_llegada = self.env.now
        servidor = self.seleccionar_servidor()
        with servidor.request() as req:
            yield req
            tiempo_espera = self.env.now - tiempo_llegada
            self.tiempos_espera.append(tiempo_espera)

            tiempo_lavado = random.expovariate(1 / self.media_lavado)  # Generar tiempo de lavado aleatorio
            yield self.env.timeout(tiempo_lavado)

            tiempo_servicio = self.generar_tiempo_servicio(servidor)
            yield self.env.timeout(tiempo_servicio)

            print(f"Auto {auto} atendido por el Servidor {servidor} en el tiempo {self.env.now}. Tiempo de espera: {tiempo_espera} min")

    def seleccionar_servidor(self):
        return random.choice(self.servidores)

    def generar_tiempo_servicio(self, servidor):
        if servidor == self.servidores[0]:
            return np.random.lognormal(mean=3, sigma=2)  # Distribución lognormal (3, 2)
        else:
            return np.random.lognormal(mean=6, sigma=1)  # Distribución lognormal (6, 1)

def llegada_autos(env, autolavado):
    i = 1
    while True:
        tiempo_llegada = random.expovariate(1 / 2)  # Tiempo de llegada de autos con tasa media de 2 autos por minuto
        yield env.timeout(tiempo_llegada)
        env.process(autolavado.lavar_auto(i))
        i += 1

env = simpy.Environment()
autolavado = Autolavado(env, num_servidores=2)
env.process(llegada_autos(env, autolavado))
env.run(until=120)  # Simular durante 20 minutos

# Gráfico de la línea de espera
tiempos = autolavado.tiempos_espera
num_autos = len(tiempos)
tiempos_espera = range(num_autos)

plt.plot(tiempos_espera, tiempos)
plt.xlabel('Número de autos')
plt.ylabel('Tiempo de espera (min)')
plt.title('Línea de espera en el autolavado')
plt.show()
```
